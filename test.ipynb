# Testing Gradience

```python
import numpy as np

from gradience.tensor import Tensor
from gradience.autograd.autograd_engine import AutogradEngine
```

### Step 1: Testing Tensor creation

```python
x = Tensor([[1, 2, 3],
            [4, 5, 6]], requires_grad=True)
print(x)
print(x.data)
print(x.shape)
print(x.dtype)
print(x.size)
print(x.requires_grad)
print(x.is_leaf)
print(x.grad)
print(x.grad_fn)
```

### Testing Add Operation.

```python
x = Tensor([1, 2, 3], requires_grad=True)
y = Tensor([4, 5, 6], requires_grad=True)

z = x + y 

print(z)
print(z.data)
print(z.is_leaf)
print(z.grad_fn)
print(z.grad_fn.operation)
print(z.grad_fn.parents)

print("-" * 80)

w = x * y

print(w)
print(w.data)
print(w.is_leaf)
print(w.grad_fn)
print(w.grad_fn.operation)
print(w.grad_fn.parents)
```
