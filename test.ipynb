# Testing Gradience

```python
import numpy as np

from gradience.tensor import Tensor
from gradience.autograd.context import Context
from gradience.ops.add import AddOp
```

### Step 1: Testing Tensor creation

```python
x = Tensor([[1, 2, 3],
            [4, 5, 6]], requires_grad=True)
print(x)
print(x.data)
print(x.shape)
print(x.dtype)
print(x.size)
print(x.requires_grad)
print(x.is_leaf)
print(x.grad)
print(x.grad_fn)
```

### Testing Operations.

```python
x = Tensor([1, 2, 3], requires_grad=True)
y = Tensor([4, 5, 6], requires_grad=True)

z = x + y 
print(z)
print(z.data)
print(z.is_leaf)
```
